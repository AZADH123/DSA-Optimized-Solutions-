**What is Bucket Sort?**

Bucket Sort is a sorting algorithm that divides numbers into buckets (groups) based on their range, sorts each bucket individually, and then combines the sorted buckets to get the final sorted list.

Steps of Bucket Sort
Create Buckets:

Divide the input range into a fixed number of buckets.
Distribute Numbers:

Place each number in its appropriate bucket based on its value.
Sort Buckets:

Sort the numbers in each bucket (using any sorting algorithm).
Combine Buckets:

Gather all numbers from the sorted buckets into one sorted list.

Advantages
Fast for Uniform Data: Efficient when numbers are evenly distributed.
Simple Implementation: Easy to understand and apply.
Disadvantages
Data Dependency: Performance depends on the distribution of data.
Extra Space: Requires additional space for buckets.
When to Use Bucket Sort?
Use Bucket Sort when:
The input data is uniformly distributed over a range.
You need to sort floating-point numbers.

```python
def bucket_sort(arr):
    if len(arr) == 0:
        return arr
    bucket_count = len(arr)
    max_value = max(arr)
    min_value = min(arr)
    bucket_range = (max_value - min_value) / bucket_count
    buckets = [[] for _ in range(bucket_count)]
    for num in arr:
        index = int((num - min_value) / bucket_range)
        if index == bucket_count:
            index -= 1
        buckets[index].append(num)
    for i in range(bucket_count):
        buckets[i].sort()
    sorted_array = []
    for bucket in buckets:
        sorted_array.extend(bucket)
    return sorted_array

if __name__ == '__main__':
    n = int(input("Enter the number of elements: "))
    arr = list(map(float, input("Enter the elements: ").split()))

    # Output
    print("Sorted Array:", bucket_sort(arr))
```
